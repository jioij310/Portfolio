```python
from pytrends.request import TrendReq   # 구글트렌드 데이터 가져오는 라이브러리
```

```python
pytrends = TrendReq(hl='ko', tz = 540)  # h1- 언어, tz- timezone(분단위설정) 우리나라 GMT+9(9시간을 분으로)
```

```python
keywords = ['sklearn', 'tensorflow', 'pytorch']
```

| 파라미터 | 설명 |
|---|---|
| `cat` | 카테고리. `0`인 경우 전체. 자세한 항목은 `pytrends.categories()`의 결과를 참고 |
| `timeframe` | 검색 시작일, 검색 종료일 |
| `geo` | 검색 지역 |
| `gprop` | 이미지, 뉴스, 유튜브, froogle(구글쇼핑)으로 세분화 할 경우 설정. 기본값은 그냥 구글 검색 |

```python
pytrends.build_payload(keywords, cat=0, timeframe= '2023-01-01 2023-12-15', geo='KR', gprop = '')
```

```python
data = pytrends.interest_over_time()
data
```

```python
pytrends.interest_by_region(
    resolution='country',   # 시, 나라, 등 지역레벨로 설정 가능
    inc_low_vol=True,  # 검색량 적은 지역 포함
    inc_geo_code=True # 국가명과 iso코드 
)
```
