```python
import datetime as dt
import requests
import os
import time
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm
import markdownify as mf
import re
```

```python
session = requests.Session()

session.headers.update({
    "Referer": "",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
})
```

```python
url = "https://magazine.hankyung.com/tag/20170213142515746"
urlFmt = "%s?page={pageNumber}" % url
maxPage = 10
```

```python


#news-tit
#lead
#date
def getContents(session, url):


      try:
            r = session.get(url)

            if r.status_code != 200:
                  msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
                  raise Exception(msg)
      except Exception as e:

            print("접속에 실패했습니다.")
            print(e)

      r.encoding = "utf-8"
      soup = BeautifulSoup(r.text)
      content = soup.select(".news-cont")
      print("추출한 컨텐츠의 수:", len(content))

      mydata = []

      for i, v in enumerate(content):
            #print(v)

            news_title = v.select(".news-tit a")
      # print(news_title)
            title = news_title[0].text.strip()
                  #print(title)
            if "href" in news_title[0].attrs:
                  href = news_title[0].attrs['href']
            
            # 수집된 주소에 도메인이 없다면 덧붙여준다.
                
            else:
                  href = None

                  #print(href)      

            lead = v.select(".lead")
            #print(lead)

            date = v.select(".date")
            datetime = date[0].text.strip()
             
            #print(date)   

            mydict = {
                  "title": title,
                  "href":href,
                  "lead":lead,
                  "date":datetime
            }
            #print(mydict)
            mydata.append(mydict)
     
      return mydata      

```

```python
news_lists = []

for i in range(1, maxPage+1):
    targetUrl = urlFmt.format(pageNumber=i)
    print(targetUrl)
    news_lists += getContents(session, targetUrl)
   #print(news_lists)

print("수집된 전체 컨텐츠 수: ", len(news_lists))  

news_lists

#print("수집된 전체 게시글 수: ", len(news_lists))
```

```python
def getContentBody(session, url):   

    try:
        r = session.get(url)
        
        if r.status_code != 200:
            msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
            raise Exception(msg)
    except Exception as e:
        print("접속에 실패했습니다.")
        print(e)

    r.encoding = "utf-8"
    soup = BeautifulSoup(r.text)
    #soup
    contentBody = soup.select(".article-body")[0]
    #print(contentBody)

    body = str(contentBody)
    #print(body)

    md = mf.markdownify(body)    

    return md
```

```python
dirname = dt.datetime.now().strftime("신문사_기사_수집_%y%m%d_%H%M%S")

if not os.path.exists(dirname):
    os.mkdir(dirname)

progress = tqdm(total=len(news_lists))

for i in news_lists:
    #print(i)
    md = getContentBody(session, i['href'])
    #print(md)

    fname = "%s/%s_%s.md" % (dirname, i['date'], i['title'])
    filename = re.sub('[\\/:*?"<>|]',"",'title')
    fname = ("%s/" % (dirname)) + fname


    print(filename)

    with open(filename, "w", encoding='utf-8') as f:
        f.write(md)

    print("%s(이)가 저장되었습니다." % fname)
    progress.update()

    time.sleep(0.1)    
```
