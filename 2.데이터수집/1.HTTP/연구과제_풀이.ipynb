# 연구과제 풀이

## 문제 1

`https://data.hossam.kr/py/titanic.json`은 타이타닉호의 탑승자 명단을 담고 있다.

이 데이터에서 `sex` 필드는 성별을 `male`과 `female`로 표현하고 있으며 탑승객의 생존 여부는 `survived`라는 이름의 boolean 형식 값으로 저장되어 있다.

남여간의 생존자와 사망자 비율을 구하라.

남자의 전체 인원을 1로 했을 때 남자 중에서의 생존자, 사망자 비율을 구하고 여자의 전체 인원을 1로 했을 때 여자 중에서의 생존자, 사망자 비율을 구해야 합니다.

### 출력예시

> 남자 생존자 비율: 0.2<br/>
> 남자 사망자 비율: 0.8<br/>
> 여자 생존자 비율: 0.7<br/>
> 여자 생존자 비율: 0.3

```python
import requests
import json
```

```python
url = "https://data.hossam.kr/py/titanic.json"
```

```python
session = requests.Session()

session.headers.update({
    "Referer": "",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
})
```

```python
try:
    r = session.get(url)
    
    if r.status_code != 200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)
except Exception as e:
    print("접속에 실패했습니다.")
    print(e)
```

```python
r.encoding = "utf-8"
mydict = json.loads(r.text)
mydict
```

```python
남자_생존 = 0
남자_사망 = 0
여자_생존 = 0
여자_사망 = 0

for i in mydict['passengers']:
    if i['sex'] == "male":
        if i['survived']:
            남자_생존 += 1
        else:
            남자_사망 += 1
    else:
        if i['survived']:
            여자_생존 += 1
        else:
            여자_사망 += 1
            
print("남자 생존: %d" % 남자_생존)
print("남자 사망: %d" % 남자_사망)
print("여자 생존: %d" % 여자_생존)
print("여자 사망: %d" % 여자_사망)
```

```python
남자_생존_비율 = 남자_생존 / (남자_생존 + 남자_사망)
남자_사망_비율 = 남자_사망 / (남자_생존 + 남자_사망)
여자_생존_비율 = 여자_생존 / (여자_생존 + 여자_사망)
여자_사망_비율 = 여자_사망 / (여자_생존 + 여자_사망)

print("남자 생존자 비율: %0.2f" % 남자_생존_비율)
print("남자 사망자 비율: %0.2f" % 남자_사망_비율)
print("여자 생존자 비율: %0.2f" % 여자_생존_비율)
print("여자 사망자 비율: %0.2f" % 여자_사망_비율)
```
