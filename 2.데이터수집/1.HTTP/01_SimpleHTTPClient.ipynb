# HTTP 통신 개요

### [1] 패키지 참조

```python
import requests
```

### [2] 접속할 데이터의 URL

```python
url = "https://www.aihub.or.kr/"
```

### [3] 세션 생성

| `세션`: 통신에서 말하는 하나의 접속 단위 혹은 접속을 수행하는 객체

```python
# 접속 객체 생성
session = requests.Session()
```

### [4] 세션에 클라이언트 정보 주입


```python
session.headers.update({
    "Referer": "",
    "User-Agent": 
"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
})
   
```

### [5] url을 통한 데이터 요청


### [1] HTTP 상태값
| 상태값 | 설명 |
|---|---|
| 200 | OK |
| 400 | 잘못된 요청 (URL 형식 에러) |
| 401 | 권한 없음 |
| 403 | 서버가 요청 거부 |
| 404 | 페이지를 찾을 수 없음 |
| 500 | 내부 서버 오류 (접속 대상 시스템의 에러) |
| 501 | 구현되지 않음 |

```python
try:
    r = session.get(url)

    if r.status_code !=200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)

except Exception as e:
    print("접속에 실패했습니다.")
    print(e)    
```

### [6] 응답 결과 확인


```python
r.encoding = "utf-8"
print(type(r.text))
print(r.text)
```
