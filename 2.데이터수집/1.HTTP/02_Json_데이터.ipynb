# JSON데이터

```python
import requests
import json
```

```python
url = "https://data.hossam.kr/py/myschool.json"
```

```python
session = requests.Session()
session.headers.update({
    "Referer": "",
    "User-Agent": 
"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
})
```

```python
try:
    r = session.get(url)

    if r.status_code !=200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        raise Exception(msg)

except Exception as e:
    print("접속에 실패했습니다.")
    print(e)  
```

```python
r.encoding = "utf-8"
print(type(r.text))
print(r.text)
```

```python
mydict = json.loads(r.text)
print(type(mydict))
mydict
```

```python
mykeys = list(mydict.keys())
mykeys
```

```python
#목적: 수집된 데이터를 분류하여 각각 저장
#방법: 반복문(헤더,내용) for문 변수 k

for k in mykeys:
    current = mydict[k]
    fname = "%s.csv" % k

    with open(fname, "w",encoding="utf-8") as f:
        for i,v in enumerate(current):
            if i == 0:
                subject = list(v.keys())
                subjectRow = "%s\n" % ",".join(subject)
                f.write(subjectRow)

            content = list(v.values())
            content = list(map(str,content))
            contentRow = "%s\n" % ",".join(content)
            f.write(contentRow)

print("fin :)")


```
