```python
# 필요한 패키지 참조

import warnings
warnings.filterwarnings(action='ignore')

import sys
import numpy as np
import seaborn as sb
from matplotlib import pyplot as plt
from pandas import read_excel
from scipy.stats import pearsonr, spearmanr
```

```python
origin = read_excel('C:/Users/Jihwan/Desktop/01Class/E.추론통계/E.InferentialStatistics/수업자료/baseball.xlsx')
origin
```

```python
plt.rcParams['font.family']='AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(12,6)
plt.rcParams['axes.unicode_minus']=False
```

```python
plt.figure()
sb.regplot(data=origin, x= '평균타율', y = '홈런수')
```

```python
# 공분산

c = np.cov(origin['홈런수'], origin['평균타율'])
c
```

```python
# numpy 상관계수

c = np.corrcoef(origin['홈런수'], origin['평균타율'])
c
```

```python
# pearson 상관계수
s, p = pearsonr(origin['홈런수'], origin['평균타율'])
'statistic: {0:.3f}, p-value: {1:.3f}, {2}가설 채택'.format(s,p,'대립' if p <= 0.05 else '귀무')
```

```python
# 데이터프레임 객체를 통한 상관계수
corr = origin.corr(method='pearson')
corr
```

```python
plt.figure(figsize=(5,3))

sb.heatmap(corr, 
           annot = True,
           cmap= 'Greens',
           vmin= -1, vmax = 1,
           linewidths=.5
        )

plt.show()
plt.close()
```

> 상관계수가 -0.075 이므로 홈런수와 평균타율는 통계적으로 유의미하다고 할 수 없다.
