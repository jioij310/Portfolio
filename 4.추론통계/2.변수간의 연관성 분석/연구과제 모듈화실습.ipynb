```python
origin = my_read_excel("https://data.hossam.kr/mldata/product.xlsx",info=False)
```

```python
origin
```

```python
data = origin.copy()
xname = "x"
yname = "y"

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))

sb.regplot(data=data, x=xname, y=yname, ax=ax1)
ax1.grid()

sb.heatmap(data.corr(method='pearson'), annot = True, cmap = 'Blues',
           vmin = -1, vmax=1, linewidths=0.5, ax=ax2)

s, p = pearsonr(data[xname], data[yname])
txt = "statistic: {0:.3f}, p-value: {1:.3f}, {2}가설 채택".format(s, p,  "대립" if p <= 0.05 else "귀무")

plt.figtext(0.51, 0, txt, wrap=True,
            verticalalignment="top", horizontalalignment="center",
            fontdict={"size": 14, "weight": "light"})

plt.show()
plt.close()
```
