```python
import pandas as pd
from pandas import read_excel

origin = read_excel('C:/Users/J/Desktop/미세먼지data/필수데이터_전처리코드/산점도_데이터(최종).xlsx')
origin

# 주어진 리스트
#a

# 리스트의 원소 중 같은 것이 있는 행 추출
df= origin[origin['역이름'].isin(a)]

# 추출된 데이터프레임 출력
empty = df.isnull()
empty
empty.sum()
df2 = df[['일평균','하차승객수']]
df2
```

```python
# 결측치 통계적 값으로 처리
from sklearn.impute import SimpleImputer
import numpy as np

imr = SimpleImputer(missing_values=np.nan, strategy='mean')
imr
df_imr = imr.fit_transform(df2.values)
df_imr
```

```python
from pandas import DataFrame
re_df2 = DataFrame(df_imr,index=df2.index, columns=df2.columns)
empty1 = re_df2.isnull()
empty1.sum()
# 결측치 없는 것 확인
re_df2
```

```python
x = re_df2['하차승객수']
z = np.polyfit(x, re_df2['일평균'], 1)
f = np.poly1d(z)
y = f(x)
y
```
