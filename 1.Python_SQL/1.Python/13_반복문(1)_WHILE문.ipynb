```python
mylist = [10, 20, 30, 40, 50]
size = len(mylist)
a = 0
while a < size:
    print(mylist[a])
    a += 1 
```

리스트 원소의 합계,평균

```python
성적표 = [92, 88 , 54, 65, 90]
과목수 = len(성적표)
i = 0
합계점수 = 0

while i < 과목수:

    합계점수 += 성적표[i]
    i += 1

    평균점수 = 합계점수 / 과목수
    print("총점: %d  평균: %0.1f" % (합계점수, 평균점수))
```

```python
a = 0
sum = 0

while a < 101:
    sum += a
    a += 2
    
print("총합:", sum)
```
