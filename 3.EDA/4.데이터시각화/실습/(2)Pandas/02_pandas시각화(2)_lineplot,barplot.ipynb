### [2] 단일 변수에 대한 시각화

DataFrame은 내부적으로 `pyplot` 객체를 초기화고 있기 때문에 그래프 초기화 메서드인 `plt.figure()`를 호출할 필요가 없다. 

그러므로 그래프의 사이즈를 설정하기 위한 `figsize` 파라미터는 그래프를 생성하는 `plot()` 메서드에서 직접 지정해야 한다.

`figsize` 파라미터를 설정하지 않을 경우 `rcParams`로 설정한 기본 크기를 따른다.

DataFrame의 index가 x축을 담당한다.

기본적으로 DataFrame의 변수 이름이 label이 되지만 `plot()` 메서드에 label 파라미터를 설정하여 다른 이름으로 변경할 수 있다.

```python
df['AUDI'].plot(label='아우디', figsize=(25,5))
plt.grid()
plt.legend()
plt.title('아우디 검색량 변동 추이')
plt.ylabel('검색량(%)')
plt.show()
plt.close()
```

### [3] 복수 변수에 대한 시각화

표시할 변수만 별도의 데이터프레임으로 추출한 후 그래프를 적용한다.

#### (1) 벤츠와 BMW만 필터링하기

```python
target = df[['BENZ','AUDI']]
target
```

#### (2) 필터링 결과를 활용한 그래프 표시

```python
target.plot(figsize=(25,5))
plt.grid()
plt.legend()
plt.title('벤츠, 아우디 검색량 변동 추이')
plt.ylabel('검색량(%)')
plt.show()
plt.close()
```

### [4] 전체 변수에 대한 시각화

```python
df.plot(figsize=(25,5))
plt.grid()
plt.legend()
plt.title('독일 자동차 3사 검색량 변동 추이')
plt.ylabel('검색량(%)')
plt.show()
plt.close()
```

### #03. Bar Chart

### [1] 데이터 가져오기

#### (1) 데이터프레임 생성

```python
df2 = read_excel('C:/Users/Jihwan/01Class/D.탐색적데이터분석/D.DataAnalysis/04.데이터시각화/수업/수업데이터/grade(1).xlsx', index_col='이름')

df2
```

#### (2) 데이터 전처리

명목형 변수에 대한 형변환

```python
df2['학년'] = df2['학년'].astype('category')
df2['성별'] = df2['성별'].astype('category')
df2.info()
```

### [2] 단일 변수에 대한 시각화

```python
df2['국어'].plot.barh(rot=0)
plt.grid()
plt.title('국어점수')
plt.show()
plt.close()
```

### [3] 복수 변수에 대한 시각화

표시할 변수만 별도의 데이터프레임으로 추출한 후 그래프를 적용한다.

각 변수의 위치를 조절하지 않더라도 범주별 막대가 겹치지 않는다.

```python
df2[['국어','영어']].plot.bar(rot=0)
plt.grid()
plt.legend()
plt.title('국어,영어 점수')
plt.show()
plt.close()
```

### [4] 전체 변수에 대한 시각화

자동으로 범주형 변수는 제외한다.

```python
df2.plot.bar(rot=0)
plt.grid()
plt.title('학생별 점수')
plt.show()
plt.close()
```
