```python
import warnings
warnings.filterwarnings(action='ignore')

import sys
from IPython.display import Image, display
from wordcloud import WordCloud

# 한글 형태소 분석기
from konlpy.tag import Okt

# 파이썬 내장 클래스 --> 설치 불필요
from collections import Counter
```

```python
nlp = Okt()
```

```python
sample = """개봉했을때부터 지금까지 마음이 답답하거나 힘들때
이영화 보고있어요 그때마다 심적인 위로를 받을수있는 영화같아요
장면 하나하나가 너무예쁘고 마음에 남아서 진한 여운까지 주는영화 감사합니다.
쩝쩝박사"""

# 명사들만 추출 -> 리스트형식으로 반환
nouns = nlp.nouns(sample)
print(nouns)
```

```python
with open('대한민국헌법.txt', 'r', encoding='utf-8') as f:
    text = f.read()
    print(text)
```

```python
nlp=Okt()
nouns = nlp.nouns(text)
print(nouns)
```

```python
with open('stopwords-ko.txt', 'r', encoding='utf-8') as f:
    stopwords = f.readlines()

    for i, v in enumerate(stopwords):
        stopwords[i] = v.strip()

    #print(stopwords)
```

```python
data_set = []

for v in nouns:
    if v not in stopwords:
        data_set.append(v)

print(data_set)
```

```python
count = Counter(data_set)
count
```

```python
wc = WordCloud(font_path="../../Fonts/malgun.ttf",
               width=1200, height=800, scale=2.0,
               background_color='#ffffff')

wc = wc.generate_from_frequencies(count)
wc.to_file('대한민국헌법-주요단어.png')
display(Image(filename='대한민국헌법-주요단어.png'))
```
