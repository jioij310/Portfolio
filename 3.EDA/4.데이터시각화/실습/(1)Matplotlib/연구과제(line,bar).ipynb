# 문제4

- 각 도시별로 2015년도의 2010년도의 인구수를 비교하는 자료를 시각화 하시오.
- 단, 인구수의 표현 범위는 200만~1000만으로 한다.

```python
from pandas import pivot_table
```

```python
origin4 = read_excel('https://data.hossam.kr/pydata/city_people.xlsx')
origin4
```

```python
df = origin4.copy()
df
df2 = df.drop([2,5])
df2 
```

```python
pivot1 = pivot_table(df2,
                     index = '도시',
                     columns='연도',
                     values='인구')

pivot1['y'] = ['200만', '600만', '1000만']
pivot1
```

```python
y1 = list(pivot1[2010])
#print(y1)
y2 = list(pivot1[2015])
#print(y2)
x = np.arange(0,len(y1))
#x,y1,y2

xticks = ['%s' % i for i in pivot1.index]
#xticks
```

```python
plt.figure()
plt.title('지역별 인구수 변화')
plt.bar(x-0.2,y1,color='#EF7C8E', label='2010',width=0.4)
plt.bar(x+0.2,y2,color='#FAE8E0', label='2015',width=0.4)
plt.xticks(x,xticks)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])
plt.xlabel('지역')
plt.ylabel('인구수')
plt.legend()
plt.grid()
plt.show()
plt.close()
```

# 문제5 

- 종사자 규모에 따른 업체 수를 확인할 수 있는 세로 막대 그래프를 구현하시오. 
- 이 때 각 데이터의 범주는 년도로 설정하시오.

```python
origin5 = read_excel('https://data.hossam.kr/pydata/jobs_by_scale.xlsx')
origin5
```

```python
y1 = list(origin5[2015])
#print(y1)
y2 = list(origin5[2016])
#print(y2)
x = np.arange(0,len(y1))
#print(x,y1,y2)

xticks = ['%s' % i for i in origin5['종사자규모별']]
#xticks
```

```python
plt.figure()
plt.title('종사자규모별 업체수')
plt.bar(x-0.2,y1,color='#EF7C8E', label='2015',width=0.4)
plt.bar(x+0.2,y2,color='#FAE8E0', label='2016',width=0.4)
plt.xticks(x,xticks)
plt.xlabel('종사자규모')
plt.ylabel('업체수')
plt.legend()
plt.grid()
plt.show()
plt.close()
```
