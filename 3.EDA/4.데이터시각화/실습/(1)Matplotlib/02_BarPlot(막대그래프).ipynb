# BarPlot(막대그래프)

## #01. 준비과정

### [1] 패키지 참조

```python
import sys
from matplotlib import pyplot as plt
import numpy as np
from pandas import DataFrame, read_excel
```

### [2] 데이터 가져오기

```python
origin = read_excel('$1', index_col='$2')
origin
```

```python
origin = read_excel('https://data.hossam.kr/pydata/grade.xlsx', index_col='이름')
origin
```

### [3] 그래프 초기화

```python
plt.rcParams['font.family']='AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(12,6)
plt.rcParams['axes.unicode_minus']=False
```

## #02. 세로 막대 그래프

### [1] 학년별 평균 데이터 구하기

```python
df = origin.drop('성별',axis=1).groupby('학년', as_index=False).mean()
df
```

### [2] 그래프에 적용할 항목 선정

```python
y = list(df['국어'])
x = list(range(0,len(y)))
xticks = ['%d학년' % i for i in df['학년']]
x, y, xticks
```

### [3] 그래프 생성

```python
plt.figure()
plt.title('학년별 국어 평균 점수')
plt.bar(x,y)
plt.xlabel('학년')
plt.ylabel('국어 평균 점수')
plt.xticks(x,xticks)
plt.grid()
plt.show()
plt.close()
```

### [4] 색상 지정하기

#### (1) 단색

```python
plt.figure()
plt.title('학년별 국어 평균 점수')
plt.bar(x,y,color='#FAE8E0')
plt.xlabel('학년')
plt.ylabel('국어 평균 점수')
plt.xticks(x,xticks)
plt.grid()
plt.show()
plt.close()
```

#### (2) 막대별로 개별 색상 지정하기


```python
plt.figure()
plt.title('학년별 국어 평균 점수')
plt.bar(x,y,color=['#EF7C8E', '#FAE8E0', '#B6E2D3', '#D8A7B1'])
plt.xlabel('학년')
plt.ylabel('국어 평균 점수')
plt.xticks(x,xticks)
plt.grid()
plt.show()
plt.close()
```

## #03. 가로 막대 그래프

### (1) 변수 생성하기

```python
x = list(df['국어'])
y = list(range(0,len(x)))
yticks = ['%d학년' % i for i in df['학년']]
x,y,yticks
```

```python
plt.figure()
plt.title('학년별 국어 평균 점수')
plt.barh(y,x,color=['#EF7C8E', '#FAE8E0', '#B6E2D3', '#D8A7B1'])
plt.ylabel('학년')
plt.xlabel('국어 평균 점수')
plt.yticks(y,yticks)
plt.grid()
plt.show()
plt.close()
```

## #04. 세로 다중 막대 그래프

### (1) 변수 생성하기


```python
y1 = list(df['국어'])
y2 = list(df['영어'])
x = np.arange(0,len(y1))
xticks = ['%d학년' % i for i in df['학년']]
y1,y2,x,xticks
```

### (2) 그래프 그리기


```python
plt.figure()
plt.title('학년별 국어, 영어 평균 점수')
plt.bar(x-0.2,y1,color='#EF7C8E',label='국어',width=0.4)
plt.bar(x+0.2,y2,color='#FAE8E0',label='영어',width=0.4)
plt.xlabel('학년')
plt.ylabel('평균 점수')
plt.xticks(x,xticks)
plt.legend()
plt.grid()
plt.show()
plt.close()
```

## #05. 가로 다중 막대 그래프

### (1) 변수 생성하기

```python
x1 = list(df['국어'])
x2 = list(df['영어'])
y = np.arange(0,len(x1))
yticks = ['%d학년' % i for i in df['학년']]
x1,x2,y,yticks
```

### (2) 그래프 그리기

```python
plt.figure()
plt.title('학년별 국어, 영어 평균 점수')
plt.barh(y-0.2,x1,color='#EF7C8E',label='국어',height=0.4)
plt.barh(y+0.2,x2,color='#FAE8E0',label='영어',height=0.4)
plt.ylabel('학년')
plt.xlabel('평균 점수')
plt.yticks(y,yticks)
plt.legend()
plt.grid()
plt.show()
plt.close()
```
