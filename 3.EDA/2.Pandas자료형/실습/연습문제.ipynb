```python
from pandas import DataFrame
import numpy as np
```

```python
data = [
        [62, 60, 63, 59, 61],
        [62, 67, 71, 64, 65, 66],
        [68, 66, 71, 67, 68, 69, 70],
        [58, 62, 60, 59, 61, 63, 64]
]
```

```python
class1 = ['A1','A2','A3','A4']
c_names = ['1회', '2회', '3회', '4회','5회','6회','7회']
df = DataFrame(data, index=class1, columns=c_names)
df.index.name = '인자의_수준'
df
```

```python
data1 = [{'A1':62},{'A1':60}, {'A1':63}, {'A1':59}, {'A1':61},
         {'A2':62},{'A2':62},{'A2':62},{'A2':62},{'A2':62},{'A2':62},
         {'A1':62}]
df1 = DataFrame(data1)
df1
```

```python
origin1 = [
        [25, 350, 425, 7],
        [27, 425, 375, 8],
        [30, 475, 485, 8],
        [35, 375, 400, 9]
]
```

```python
origin2 = [
        [20, 250, 300, 7],
        [33, 325, 225, 9],
        [37, 300, 375, 9]
        ]
```
