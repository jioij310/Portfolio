# 재구조화_피벗테이블, melt

```python
from pandas import read_excel, pivot_table, melt, DataFrame
```

```python
origin = read_excel("https://data.hossam.kr/pydata/city_people.xlsx")
origin
```

```python
pivot1 = pivot_table(origin,
                     index = ['지역','도시'],
                     columns='연도',
                     values='인구')
print(type(pivot1))
pivot1
```

```python
pivot_table(origin,
            index = '지역',
            columns= '연도',
            values='인구',
            aggfunc='sum'

            )
```

```python
pivot_table(origin,
            index = '지역',
            columns= '연도',
            values='인구',
            aggfunc=['sum', 'mean']

            )
```

```python
pivot_table(origin,
            index = ['지역','연도'],
            columns= '도시',
            values='인구',
            aggfunc=['sum','mean']
           )
```

## melt

```python
pivot_df = pivot_table(origin,
            index = '연도',
            columns= '지역',
            values='인구',
            aggfunc='mean'

            )

pivot_df
```

```python
pivot_df2 = pivot_df.reset_index()
pivot_df2
```

```python
mdf = melt(pivot_df2, id_vars=['연도'], value_vars=['경상권','수도권'])
mdf
```

```python
mdf = melt(pivot_df2, id_vars = ['연도'], value_vars = ['경상권', '수도권'], var_name = '구분', value_name='인구수')
mdf
```

### 그룹별 집계

```python
origin = read_excel("https://data.hossam.kr/pydata/traffic_acc.xlsx")
origin
```

연도별 합계 

```python
origin.groupby('년도').sum()
```

```python
df = origin.drop('월',axis=1)
df
```

```python
df.groupby('년도').sum()
```

```python
df = df.filter(['년도', '사망자수'])
df
```

```python
df.groupby('년도').sum()
```

```python
origin.groupby(['년도', '월']).mean()
```

```python
origin.filter(['년도', '발생건수']).groupby('년도').agg(['sum', 'max', 'min', 'mean'])
```

```python
def myFunction(x):
    return x.max() - x.min()
```

```python
origin.drop('월',axis=1).groupby('년도').agg(['max', 'min',lambda x: x.max() - x.min()])
```

```python
origin.drop('월',axis = 1).groupby('년도').agg(['max', 'min', myFunction])
```

```python
origin.drop('월', axis=1).groupby('년도',as_index=False).sum()
```

컬럼별로 다른 집계함수 사용하는 경우

```python
origin.drop("월", axis = 1).groupby('년도').agg(
    {'발생건수': ['mean', 'sum'], '사망자수':'max', '부상자수':'min'}
)
```

데이터 정규화

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
origin = read_excel('https://data.hossam.kr/pydata/gradeuate.xlsx')
origin
```

```python
df = origin.copy()

xmin = df['필기점수'].min()
xmax = df['필기점수'].max()

df['필기점수_MinMax(1)'] = (df['필기점수']-xmin) / (xmax - xmin)
df
```

```python
n = df['필기점수'].values
n
```

```python
n_re = n.reshape(-1,1)
n_re
```

```python
scaler = MinMaxScaler()
scaler.fit(n_re)
df['필기점수_MinMax(2)']=scaler.transform(n_re)
df
```

```python
tmp= df.filter(['필기점수'])
tmp
```

```python
df[['필기점수']]
```

```python
scaler = MinMaxScaler()
scaler.fit(tmp)
df['필기점수_MinMax(3)'] = scaler.transform(tmp)
df
```

```python
scaler= MinMaxScaler()
scaler.fit(origin)
mms = scaler.transform(origin)
mms
```

```python
minmax_df= DataFrame(mms, index=origin.index, columns= origin.columns)
minmax_df
```

```python
평균=df['학부성적'].mean()
표준편차=df['학부성적'].std()
df['학부성적_Standard(1)']=(df['학부성적']-평균)/표준편차
df
```

```python
n = df['학부성적'].values.reshape(-1,1)
n
```

```python
scaler = StandardScaler()
scaler.fit(n)
df['학부성적_Standard(2)']=scaler.transform(n)
df
```

```python
중앙값 = df['병원경력'].median()
iqr= df['병원경력'].quantile(0.75) - df['병원경력'].quantile(0.25)
df['병원경력_Robust(1)'] = (df['병원경력']-중앙값) / iqr
df
```

```python
n = df['병원경력'].values.reshape(-1,1)
scaler = RobustScaler()
scaler.fit(n)
df['병원경력_Robust(2)']=scaler.transform(n)
df
```
